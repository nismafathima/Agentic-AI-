```python

!pip install -qU langchain-community pypdf
from langchain_community.document_loaders import PyPDFLoader  # Fixed import
loader = PyPDFLoader(f"/content/{list(file_name)[0]}")
document = loader.load()
```

```python

!pip install -qU langchain-text-splitters
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
Chunks=splitter.split_documents(document)
```

```python

!pip install -qU langchain-huggingface
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embeddings=HuggingFaceEmbeddings(
model="sentence-transformers/all-MiniLM-L6-v2")
```

```python

!pip install -qU faiss-cpu  # Also need FAISS itself
from langchain_community.vectorstores import FAISS
```

```python
vector_db=FAISS.from_documents(Chunks, embeddings)
```

```python

# Install ALL packages correctly
!pip install -qU langchain-community langchain-huggingface langchain-text-splitters faiss-cpu transformers torch pypdf

from langchain_community.llms import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
```

```python
model_name="gpt2" # Changed from "gbt-2" to "gpt2" to fix the model not found error
tokenizer=AutoTokenizer.from_pretrained(model_name)
model=AutoModelForCausalLM.from_pretrained(model_name)

pipe = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    max_new_tokens=60,
    do_sample=True,
    temperature=0.7,
    pad_token_id=tokenizer.eos_token_id
)
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
# AGENTIC AI - Works with your llm + vector_db
class AgenticAI:
    def __init__(self, llm, vector_db):
        self.llm = llm
        self.vector_db = vector_db
        self.tasks_done = []

    def plan_mission(self, goal):
        """1. PLAN: Breaks goal into steps"""
        plan_prompt = f"""
        GOAL: {goal}
        Available: PDF analysis tools

        Make 3-step PLAN:
        1. [Step 1]
        2. [Step 2]
        3. [Step 3]"""
        plan = self.llm.invoke(plan_prompt)
        print("ğŸ“‹ PLAN:", plan)
        return plan

    def execute_task(self, task):
        """2. EXECUTE: Uses your RAG to do tasks"""
        # Your PDF becomes the agent's "knowledge base"
        docs = self.vector_db.similarity_search(task, k=3)
        context = "\n".join([doc.page_content for doc in docs]) # Fixed syntax error here

        action_prompt = f"""TASK: {task}
        PDF Context: {context}

        DO THE TASK:"""
        result = self.llm.invoke(action_prompt)
        self.tasks_done.append(result)
        print("âœ… DONE:", result)
        return result

    def check_progress(self, goal, results):
        """3. REFLECT: Did we succeed?"""
        reflect_prompt = f"""
        GOAL: {goal}
        RESULTS: {results}

        SUCCESS? Next step or FINISHED?"""
        reflection = self.llm.invoke(reflect_prompt)
        print("ğŸ’­ THINKING:", reflection)
        return "FINISHED" in reflection

    def run_mission(self, goal):
        print(f" MISSION: {goal}")

        # Step 1: Plan
        plan = self.plan_mission(goal)

        # Step 2: Execute 3 tasks
        tasks = ["analyze content", "find key points", "summarize findings"]
        all_results = []

        for task in tasks:
            result = self.execute_task(task)
            all_results.append(result)

        # Step 3: Final reflection
        self.check_progress(goal, all_results)

        return all_results

# RUN YOUR AGENT:
agent = AgenticAI(llm, vector_db)
agent.run_mission("Understand .pdf completely")
```
